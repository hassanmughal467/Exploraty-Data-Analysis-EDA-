```python
df_country=pd.read_excel("./Country-Code.xlsx")
df_country.head()
```

```python
df.columns
```

```python
final_df=pd.merge(df,df_country,on="Country Code",how="left")
```

```python
final_df
```

```python
# to check data types
final_df.dtypes
```

```python
country_names=final_df.Country.value_counts().index
```

```python
country_values=final_df.Country.value_counts().values
```

```python
#pie chart to top 3 country
plt.pie(country_values[:3],labels=country_names[:3],autopct="'%1.2f%%'")
```

###### Observation:Zomato maximum records or transaction are from India After that USA and then United Kingdoms

```python
final_df.columns
```

```python
ratings=final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
ratings
```

#### Observation
1. When Rating is between 4.5 to 4.9---> Excellent
2. When Rating are between 4.0 to 3.4--->very good
3. when Rating is between 3.5 to 3.9----> good
4. when Rating is between 3.0 to 3.4----> average
5. when Rating is between 2.5 to 2.9----> average
6. when Rating is between 2.0 to 2.4----> Poor

```python
ratings.head()
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize'] = (12, 6)
sb.barplot(x="Aggregate rating",y="Rating Count",data=ratings)
```

```python
sb.barplot(x="Aggregate rating",y="Rating Count",hue='Rating color',data=ratings,palette=['blue','red','orange','yellow','green','green'])
```

#### Observation:

1. Not Rated count is very high
2. Maximum number of rating are between 2.5 to 3.4

```python
## Count plot
sb.countplot(x="Rating color",data=ratings,palette=['blue','red','orange','yellow','green','green'])
```
